```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/Customer_Care_Stats_2021___2023_FCC.csv')

# Preprocess numeric columns to remove commas and convert to float
numeric_columns = ['Calls', 'Emails', 'Footfall']
for col in numeric_columns:
    df[col] = df[col].str.replace(',', '').astype(float)

# Time Series Plot with Trendlines
plt.figure(figsize=(12, 6))
plt.plot(df['Month'], df['Calls'], marker='o', color='blue', label='Calls')
plt.plot(df['Month'], df['Emails'], marker='o', color='green', label='Emails')
plt.plot(df['Month'], df['Footfall'], marker='o', color='orange', label='Footfall')
plt.xlabel('Month')
plt.ylabel('Count')
plt.title('Call Center Activity Over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Bar Chart for Call Volume Comparison
plt.figure(figsize=(12, 6))
plt.bar(df['Month'], df['Calls'], color='skyblue', label='Calls')
plt.xlabel('Month')
plt.ylabel('Number of Calls')
plt.title('Call Volume by Month')
plt.xticks(rotation=45)
plt.grid(axis='y')
# Add annotations for important events if the 'Notes' column exists
if 'Notes' in df.columns:
    for i, event in enumerate(df['Notes'].dropna()):
        plt.text(i, df['Calls'].max() * 0.9, event, rotation=45, ha='center', va='center', fontsize=8, color='red')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv('/content/Customer_Care_Stats_2021___2023_FCC.csv')

# Preprocess numeric columns to remove commas and convert to float
numeric_columns = ['Calls', 'Emails', 'Footfall']
for col in numeric_columns:
    df[col] = df[col].str.replace(',', '').astype(float)

# Stacked Bar Chart for Composition of Calls by Month
df_monthly = df[df['Month'].str.lower().str.contains('total') == False]  # Exclude total rows
df_monthly.set_index('Month')[['Calls', 'Emails']].plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Month')
plt.ylabel('Number of Interactions')
plt.title('Composition of Calls and Emails by Month')
plt.xticks(rotation=45)
plt.legend(title='Interaction Type')
plt.tight_layout()
plt.show()

# Heatmap for Call Volume by Month and Day of Week
df_heatmap = df.pivot_table(index='Month', columns='Day', values='Calls', aggfunc='sum')
plt.figure(figsize=(12, 6))
sns.heatmap(df_heatmap, cmap='Blues', annot=True, fmt='g', linewidths=0.5)
plt.xlabel('Day of Week')
plt.ylabel('Month')
plt.title('Call Volume by Month and Day of Week')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Box Plot for Call Duration
# Assuming there's a column named 'Call_Duration' for call duration
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Month', y='Call_Duration')
plt.xlabel('Month')
plt.ylabel('Call Duration')
plt.title('Distribution of Call Duration by Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
